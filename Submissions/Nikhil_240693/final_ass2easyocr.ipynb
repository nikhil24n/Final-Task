```python
import easyocr
```

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = "/content/IMG_5152.PNG"  # Corrected path

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_path1 = "/content/IMG_5154.JPG"
img1= cv2.imread(img_path1)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.show()
plt.imshow(img1)
plt.axis("off")
plt.show()
```

```python
reader = easyocr.Reader(['en'])
```

```python
output = reader.readtext('/content/IMG_5152.PNG')
output
```

```python
output1 = reader.readtext('/content/IMG_5154.JPG')
output1
```

```python
!pip install jiwer
from jiwer import wer, cer
```

```python
predicted = " ".join([item[1] for item in output])
actual = "This is a handwritten example Write as good as you can."
print(predicted)
wer_score = wer(actual, predicted)
cer_score = cer(actual, predicted)

print("WER:", wer_score)
print("CER:", cer_score)
```

```python
predicted1 = " ".join([item[1] for item in output1])
actual1 = "57"
wer_score = wer(actual1, predicted1)
cer_score = cer(actual1, predicted1)

print("WER1:", wer_score)
print("CER1:", cer_score)

```
